Since the cv.imshow is making the notebook to hang and the terminal dies, so we will write a function as we need to call these few lines again and again, if we want to display images using the matplotlib, so we will write a wrapper and get the display code in this function and call it when we need to show an image.

```python
def display_image(image_pixels):
    plt.figure(figsize=(8,8))
    plt.imshow(image_pixels)
    plt.axis("off")
    plt.show()
```

Let us now see how the drawing is done in the cv2, like drawing the circles, rectangles etc over the images.
Let us first create the image 500px and 500px containing all 0's. As the image is pixel array, so we can create a black or whote image using the 0 or 255 pixel values with the shape 500 500.

```python
import numpy as np
def generate_blk(x,y):
    image = np.zeros((x, y, 3), dtype=np.uint8)
    display_image(image)
    return image
```

```python
generate_blk(500,500)
```

Now as we have generated the black image, we will now see how to draw the rectangles, lines and circles over this empty canvas. This canvas could be an image also, which we will see towards the end.

```python
canvas = generate_blk(500,500)
```

Let us see how can we draw the line. We will need the following command `cv2.line(canvas,startXY, endXY, color, sizeStroke)`

```python
green = (0,255,0)
cv2.line(canvas, (0,0), (250,250), green)
display_image(canvas)
```

We can aslo give the size to the stroke

```python
green = (0,255,0)
cv2.line(canvas, (0,0), (250,250), green, 4)
display_image(canvas)
```

let us see how can we now get the rectangle on the canvas. One thing to keep in mind is that we need to get strated is we need to give two x, y's. If we want our ABCD rectangle, we need to give the co ordinates of A and C. pic.

The function for the rectangle is: `cv2.rectange(canvas, A, C, color, Size)`

```python
# b g r
red = (255, 0, 0)
cv2.rectangle(canvas, (10,10), (30,300), red, 3)
display_image(canvas)
```

In the similar fashion, we can draw circle `cv2.circle(canvas, centerXY, radius, color, size)`

```python
blue = (0, 0, 255)
cv2.circle(canvas, (250,250), 220, blue, 3)
display_image(canvas)
```

Now for drawing over the images, all we need to do is load the image and pass that image to the functions we just saw.

```python
image = cv2.imread("example.jpg")
display_image(image)
```

```python
cv2.rectangle(image, (150, 150), (300,250), red, 3)
display_image(image)
```
